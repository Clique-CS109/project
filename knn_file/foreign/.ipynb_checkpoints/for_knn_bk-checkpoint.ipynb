```python
traindf = pd.read_csv("for_train.csv", dtype={'ISBN':np.str, 'ID':np.str, 'Date':np.str, 'Pub_Date':np.str, 'Cart_Date':np.str})
validatedf = pd.read_csv("for_validate.csv", dtype={'ISBN':np.str,'ID':np.str, 'Date':np.str,'Pub_Date':np.str,'Cart_Date':np.str})
testdf = pd.read_csv("for_test.csv", dtype={'ISBN':np.str,'ID':np.str,'Date':np.str,'Pub_Date':np.str,'Cart_Date':np.str})
```

```python
# delete [염가한정판매] from title

def del_yeomga(df):
    resdf=df.copy()
    title_series=resdf['Title'].apply(lambda r: r if r[:21]!='[염가한정판매] ' else r[21:])
    resdf['Title']=title_series
    return resdf

traindf=del_yeomga(traindf)
validatedf=del_yeomga(validatedf)
testdf=del_yeomga(testdf)
```

```python
# drop duplicates - I will NOT use "Count" column
traindf=traindf.drop_duplicates(['ID','ISBN'])
validate=validatedf.drop_duplicates(['ID','ISBN'])
testdf=testdf.drop_duplicates(['ID','ISBN'])
```

```python
# drop if ISBN==nan
traindf=traindf[~pd.isnull(traindf['ISBN'])]
validatedf=validatedf[~pd.isnull(validatedf['ISBN'])]
testdf=testdf[~pd.isnull(testdf['ISBN'])]
```

```python
def recompute_frame(ldf):
    ldf_user=ldf.groupby('ID')
    ldf_book=ldf.groupby('ISBN')
    
    user_purchase_count=ldf_user.size()
    book_sell_count=ldf_book.size()

    nldf=ldf.copy()
    
    nldf.set_index(['ISBN'], inplace=True)
    nldf['book_sell_count']=book_sell_count
    nldf.reset_index(inplace=True)
    
    nldf.set_index(['ID'], inplace=True)
    nldf['user_purchase_count']=user_purchase_count
    nldf.reset_index(inplace=True)
    
    return nldf
```

```python
def make_smaller(df, bookcount, usercount):
    smallidf1=df[(df.book_sell_count > bookcount)]
    smallidf1=recompute_frame(smallidf1)
    smallidf2=smallidf1[(smallidf1.user_purchase_count > usercount)]
    smalldf=recompute_frame(smallidf2)
    return smalldf
```

```python
traindf=recompute_frame(traindf)
validatedf=recompute_frame(validatedf)
testdf=recompute_frame(testdf)
```

```python
def make_vec(items,total):
    res=np.zeros(len(total), dtype=bool)
    for i in items:
        res[np.argwhere(total==i)[0][0]]=True
    return res


def compute_supports_for_book(df):
    IDs=df.ID.unique()
    gb_ISBN = df.groupby('ISBN').ID.unique()
    res={}
    for e,v in zip(gb_ISBN.index.values, gb_ISBN.values):
        res[e] = make_vec(v,IDs)
    return res

def compute_supports_for_customer(df):
    ISBNs = df.ISBN.unique()
    gb_ID = df.groupby('ID').ISBN.unique()
    res={}
    i=0
    for e,v in zip(gb_ID.index.values, gb_ID.values):
        res[e] = make_vec(v,ISBNs)
        i+=1
        if i%1000==0:
            print i
    return res


def hist_data(bdict):
    books=bdict.keys()
    supports=[]
    for i,book1 in enumerate(books):
        for j,book2 in enumerate(books):
            if  i < j:
                supmask = (bdict[book1] & bdict[book2])
                common_buyers = np.sum(supmask)
                supports.append(common_buyers)
    print "mean support",np.mean(supports), "median support", np.median(supports)
    return supports
```

# Find kNN -------------------------------------------------------------------------

```python
def find_knn(df, cid):
    clist=set(df[df.ID==cid].ISBN.unique())
    res={}
    sup_num=len(clist)
    
    if sup_num==0:
        return {}
    
    for i, v in df.groupby('ID').ISBN.unique().iteritems():
        if i != cid:
            int_num=len(clist.intersection(set(v)))
            if int_num !=0:
                res[i]=int_num*1./(np.sqrt(sup_num*len(set(v))))
    return res
```

```python
def get_result(df, knn_dist, cvlist, k=5):
    knn=knn_dist[:k]
    res=np.zeros(len(df.ISBN.unique()))
    for (i,j) in knn:
        res+=cvlist[i]*j/np.sqrt(np.sum(cvlist[i]))
    return {k:v for (k,v) in zip(df.ISBN.unique(),res)}

def get_allbooks_prob(df):
    temp=df.groupby('ISBN').ID.unique().apply(len)
    return dict(temp.apply(lambda r: r*1./temp.sum()))

def get_bought_list(df, cid):
    res={i:0 for i in df.ISBN.unique()}
    for i in df[df.ID==cid].ISBN.unique():
        res[i]=-np.inf
#     return {i:(-np.inf if i in df[df.ID==cid].ISBN.unique() else 0) for i in df.ISBN.unique()}
    return res
```

```python
def do_grid_search(df, testset, cid, iter_k, iter_p, found_supports, allbooks_prob, found_knn, already_bought):
    numlist={}
    for k in iter_k:
        knnresult=get_result(df, found_knn, found_supports,k=k)
        for p in iter_p:
            if (k,p) in all_grid_data:
                numlist[(k,p)]=all_grid_data[(k,p)]
            else:
                res={book:(allbooks_prob[book]+already_bought[book]+knnresult[book]*p) for book in df.ISBN.unique()}
                numlist[(k,p)]=len(set(sorted(res, key=res.get, reverse=True)[:5]).intersection(testset))
    return numlist
# for i in np.random.choice(validatedf.ID.unique(),10, replace=False):
#     print do_grid_search(traindf,validatedf, i, iter_k, iter_p, found_supports, allbooks_prob)
```

```python
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import cm

def plot_grid_result(plotdata, iter_k, iter_p):
    X=np.outer(iter_k,np.ones(len(iter_p),dtype=np.int))
    Y=np.outer(np.ones(len(iter_k)), iter_p) #x,y for 3D plots
    Z=[] # these are for 3D wire plot         
    for k in iter_k:
        Z.append([plotdata[(k,p)] for p in iter_p])
    
    ax=Axes3D(plt.figure())
    ax.plot_surface(X,Y,Z,rstride=1,cstride=1,cmap=cm.coolwarm)
    ax.view_init(70, 20) # 3D plot
    ax2=Axes3D(plt.figure())
    ax2.plot_wireframe(X,Y,Z)
    ax2.view_init(20, 70)
```
